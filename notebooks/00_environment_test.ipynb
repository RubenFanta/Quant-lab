```python
import numpy as np
import pandas as pd
import yfinance as yf
```

```python
aapl = yf.download(
    "AAPL",
    start="2025-01-01",
    end="2026-01-01",
    progress=False,
auto_adjust=False
)
```

```python
prices = aapl["Close"]

if isinstance(prices, pd.DataFrame):
    prices = prices["AAPL"]
```

```python
df = pd.DataFrame()
df["Close"] = prices
```

```python
df["simple_return"] = prices.pct_change()
df["simple_return_pct"] = df["simple_return"]*100
yearly_return_from_prices = prices.iloc[-1]/prices.iloc[0]-1
```

```python
df.head
```

```python
print(yearly_return_from_prices)
print(yearly_return_from_prices*100)
```

```python
df["log_return"] = np.log(prices/prices.shift(1))
df["log_return_%"] = df["log_return"]*100
df["simple_min_log"] = df["simple_return_pct"] - df["log_return_%"]
```
